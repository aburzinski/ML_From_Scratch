### Import Packages

```python
import numpy as np
from sklearn import datasets
```

### Create Dataset

```python
dataset = datasets.load_diabetes()
y = dataset.target
X = dataset.data

# X = np.array([[1, 1],
#               [2, 1],
#               [2, 2],
#               [3, 4]])

# y = np.array([2, 3, 4, 6])

X = np.insert(X, 0, 1, axis = 1)
```

### Solve for Beta_hats

```python
Xt = np.transpose(X)
Xt_X_inv = np.linalg.inv(np.dot(Xt, X))

beta_hat = np.linalg.multi_dot([Xt_X_inv, Xt, y])

beta_hat
```

### Compute Predictions

```python
y_hat = np.dot(X, beta_hat)

y_hat
```

### Compute RMSE

```python
rmse = np.sqrt(np.sum((y - y_hat) ** 2) / y.shape[0])

rmse
```

### Hat Matrix

```python
Xt = np.transpose(X)
Xt_X_inv = np.linalg.inv(np.dot(Xt, X))

H = np.linalg.multi_dot([X, Xt_X_inv, Xt])

np.dot(H, y)
```

### Standard Error of Betas

```python
rmse = np.sqrt(np.sum((y - y_hat) ** 2) / y.shape[0])
Xt_X_inv = np.linalg.inv(np.dot(Xt, X))

var_beta = rmse * Xt_X_inv
np.diagonal(var_beta)
```
